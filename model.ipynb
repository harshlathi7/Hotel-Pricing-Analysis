# Hotel-Pricing-Analysis
## Machine Learning Models

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
df_mum = pd.read_csv("Hotel-Pricing-Analysis/Dataset/Cleaned_Score/mumbai_hotels_scored_cleaned.csv")
df_mum.insert(2, "City", ['Mumbai' for x in range(df_mum.shape[0])], True)
df_del = pd.read_csv("Hotel-Pricing-Analysis/Dataset/Cleaned_Score/delhi_hotels_scored_cleaned.csv")
df_del.insert(2, "City", ['Delhi' for x in range(df_del.shape[0])], True)
df_ban = pd.read_csv("Hotel-Pricing-Analysis/Dataset/Cleaned_Score/bangalore_hotels_scored_cleaned.csv")
df_ban.insert(2, "City", ['Bangalore' for x in range(df_ban.shape[0])], True)
df_che = pd.read_csv("Hotel-Pricing-Analysis/Dataset/Cleaned_Score/chennai_hotels_scored_cleaned.csv")
df_che.insert(2, "City", ['Chennai' for x in range(df_che.shape[0])], True)
df_kol = pd.read_csv("Hotel-Pricing-Analysis/Dataset/Cleaned_Score/kolkata_hotels_scored_cleaned.csv")
df_kol.insert(2, "City", ['Kolkata' for x in range(df_kol.shape[0])], True)

df = df_mum.copy(deep=True)
df = df.append(df_del, ignore_index = True)
df = df.append(df_ban, ignore_index = True)
df = df.append(df_che, ignore_index = True)
df = df.append(df_kol, ignore_index = True)
dum = pd.get_dummies(df.City)
df=df.drop('City', axis=1)
df = pd.concat([df, dum], axis=1)
df
```

## Normalization (min-max normalization)

```python
# Function to normalize the price
def normalize(in_price):
    return (in_price-Min['Price'])/(Max['Price']-Min['Price'])

# Fuction to denormalize the price
def denormalize(pred_price):
    return pred_price*(Max['Price']-Min['Price'])+Min['Price']
```

```python
df_norm = df.copy(deep=True)
Min = {}
Max = {}

Min['Latitude'] = min(df_norm['Latitude'])
Min['Longitude'] = min(df_norm['Longitude'])
Min['Rating'] = min(df_norm['Rating'])
Min['Number of reviews'] = min(df_norm['Number of reviews'])
Min['Price'] = min(df_norm['Price'])

Max['Latitude'] = max(df_norm['Latitude'])
Max['Longitude'] = max(df_norm['Longitude'])
Max['Rating'] = max(df_norm['Rating'])
Max['Number of reviews'] = max(df_norm['Number of reviews'])
Max['Price'] = max(df_norm['Price'])

df_norm['Latitude'] = (df_norm['Latitude'] - min(df_norm['Latitude']))/(max(df_norm['Latitude']) - min(df_norm['Latitude']))
df_norm['Longitude'] = (df_norm['Longitude'] - min(df_norm['Longitude']))/(max(df_norm['Longitude']) - min(df_norm['Longitude']))
df_norm['Rating'] = (df_norm['Rating'] - min(df_norm['Rating']))/(max(df_norm['Rating']) - min(df_norm['Rating']))
df_norm['Number of reviews'] = (df_norm['Number of reviews'] - min(df_norm['Number of reviews']))/(max(df_norm['Number of reviews']) - min(df_norm['Number of reviews']))
df_norm['Price'] = (df_norm['Price'] - min(df_norm['Price']))/(max(df_norm['Price']) - min(df_norm['Price']))
df_norm
```

```python
X = df_norm.drop(['Name', 'Price'], axis = 1)
X = np.array(X)
```

```python
y = np.array(df_norm['Price'])
```

## Feature selection using RFE (Recursive Feature Elimination)

```python
# Feature selection
Model = LinearRegression()
Rfe = RFE(Model, n_features_to_select = 50)
Fit = Rfe.fit(X,y)
selected = Fit.support_
selected
```

```python
tbd = []
l = len(selected)
for i in range(l):
    if selected[i]==False:
        tbd.append(i)
        
tbd
```

```python
X = np.delete(X, tbd, 1)
X
```

## Splitting of data into train and test set

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=53)
# for random state = 42 and number of features = 25 the results are good
# also for random state = 40 and number of features = 50, test_size = 0.2 in both cases
# random_state = 40
# random_state = 59 for scores (or 18 which is less)
# random_state = 53 for all cities together
```

## Linear Regression Model

```python
reg = LinearRegression().fit(X_train,y_train)
```

```python
print("train score =", reg.score(X_train,y_train))
print("test score =", reg.score(X_test,y_test))
```

## Finding the optimal value of number of features to be selected by RFE for highest score

```python
mx_test = 0
mx_train = 0
for j in range(58):
    X = df_norm.drop(['Name', 'Price'], axis = 1)
    X = np.array(X)
    y = np.array(df_norm['Price'])
    Model = LinearRegression()
    Rfe = RFE(Model, n_features_to_select = j+1)
    Fit = Rfe.fit(X,y)
    selected = Fit.support_
    tbd = []
    l = len(selected)
    for i in range(l):
        if selected[i]==False:
            tbd.append(i)
    X = np.delete(X, tbd, 1)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=53)
    reg = LinearRegression().fit(X_train,y_train)
    print(j+1, "train score =", reg.score(X_train,y_train))
    print(j+1, "test score =", reg.score(X_test,y_test))
    if mx_test < reg.score(X_test,y_test):
        mx_test = reg.score(X_test,y_test)
        test_train = reg.score(X_train,y_train)
        pos_test = j+1
    if mx_train < reg.score(X_train,y_train):
        mx_train = reg.score(X_train,y_train)
        train_test = reg.score(X_test,y_test)
        pos_train = j+1
print("\nHighest train score =",mx_train)
print("Corressponding test score =",train_test)
print("For",pos_train,"selected features")
print("\nHighest test score =",mx_test)
print("Corressponding test score =",test_train)
print("For",pos_test,"selected features")
```

##  Random Forest Regression model

```python
X = df_norm.drop(['Name', 'Price'], axis = 1)
X = np.array(X)
y = np.array(df_norm['Price'])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=53)
```

```python
rfr = RandomForestRegressor(n_estimators=100)
rfr_model = rfr.fit(X_train,y_train)
```

```python
print("test score =", rfr_model.score(X_test,y_test))
print("train score =", rfr_model.score(X_train,y_train))
```
