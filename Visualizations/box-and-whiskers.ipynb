```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_del = pd.read_csv("Dataset/delhi_hotels.csv")
df_mum = pd.read_csv("Dataset/mumbai_hotels.csv")
df_ban = pd.read_csv("Dataset/bangalore_hotels.csv")
df_che = pd.read_csv("Dataset/chennai_hotels.csv")
df_kol = pd.read_csv("Dataset/kolkata_hotels.csv")
```

```python
# Electricity generation from various sources for all the countries

D = []

data = []
for i in df_del.index:
    data.append(df_del.iloc[i]['Price'])

D.append(data)
data = []
for i in df_mum.index:
    data.append(df_mum.iloc[i]['Price'])

D.append(data)
data = []
for i in df_ban.index:
    data.append(df_ban.iloc[i]['Price'])

D.append(data)
data = []
for i in df_che.index:
    data.append(df_che.iloc[i]['Price'])

D.append(data)
data = []
for i in df_kol.index:
    data.append(df_kol.iloc[i]['Price'])

D.append(data)

f = plt.figure(figsize=(10,10))
a = f.add_subplot(111)
a.set_xticklabels(['Delhi', 'Mumbai', 'Bangalore', 'Chennai', 'Kolkata'])
a.set_title('Hotel prices (in Rupees) accross cities')
plt.boxplot(D)
plt.show()
```
