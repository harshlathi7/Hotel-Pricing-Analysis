```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.DataFrame(pd.read_csv("Dataset/Cleaned Data/bangalore_hotels_cleaned.csv"))
df2 = pd.DataFrame(pd.read_csv("Dataset/Cleaned Data/delhi_hotels.csv"))
df3 = pd.DataFrame(pd.read_csv("Dataset/Cleaned Data/mumbai_hotels.csv"))
df4 = pd.DataFrame(pd.read_csv("Dataset/Cleaned Data/chennai_hotels.csv"))
df5 = pd.DataFrame(pd.read_csv("kolkata_hotels.csv"))
```

```python
df1["sum"] = df1.iloc[:, 3:50].sum(axis=1)
df2["sum"] = df2.iloc[:, 3:50].sum(axis=1)
df3["sum"] = df3.iloc[:, 3:50].sum(axis=1)
df4["sum"] = df4.iloc[:, 3:50].sum(axis=1)
df5["sum"] = df5.iloc[:, 3:50].sum(axis=1)
```

```python
x1 = df1["Price"]
y1 = df1["sum"]

x2 = df2["Price"]
y2 = df2["sum"]

x3 = df3["Price"]
y3 = df3["sum"]

x4 = df4["Price"]
y4 = df4["sum"]

x5 = df5["Price"]
y5 = df5["sum"]

plt.scatter(x1,y1,c="red")
plt.scatter(x2,y2,c="green")
plt.scatter(x3,y3,c="blue")
plt.scatter(x4,y4,c="yellow")
plt.scatter(x5,y5,c="black")

plt.xlabel("Price in Rs.")
plt.ylabel("Number of services")

plt.show()
```

```python
x = ['Bangalore', 'Delhi', 'Mumbai', 'Chennai', 'Kolkata']
y = [df1.shape[0], df2.shape[0], df3.shape[0], df4.shape[0], df5.shape[0]]

plt.bar(x, y, color="maroon", width=0.5)
plt.xlabel("Cities")
plt.ylabel("Number of hotels")
plt.show()
```

```python
price = df1['Price']
price = price.append(df2['Price'])
price = price.append(df3['Price'])
price = price.append(df4['Price'])
price = price.append(df5['Price'])

price = price.sort_values()
hotel = [ i for i in range(1,len(price)+1)]

plt.scatter(hotel, price, c="blue")
plt.xlabel("Hotels")
plt.ylabel("Prices")
plt.show()
```
