```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from scipy.stats import norm
```

```python
df1 = pd.DataFrame(pd.read_csv("Dataset/Cleaned_Score/bangalore_hotels_scored_cleaned.csv"))
df2 = pd.DataFrame(pd.read_csv("Dataset/Cleaned_Score/delhi_hotels_scored_cleaned.csv"))
df3 = pd.DataFrame(pd.read_csv("Dataset/Cleaned_Score/mumbai_hotels_scored_cleaned.csv"))
df4 = pd.DataFrame(pd.read_csv("Dataset/Cleaned_Score/chennai_hotels_scored_cleaned.csv"))
df5 = pd.DataFrame(pd.read_csv("Dataset/Cleaned_Score/kolkata_hotels_scored_cleaned.csv"))
```

```python
x = ['Bangalore', 'Delhi', 'Mumbai', 'Chennai', 'Kolkata']
y = [df1.shape[0], df2.shape[0], df3.shape[0], df4.shape[0], df5.shape[0]]

df = pd.DataFrame({"City":x, "Hotels":y})
df = df.sort_values('Hotels', ascending=False)

plt.figure(figsize=(10,6))
plt.bar('City', 'Hotels', data=df)
plt.xlabel("City", size=15)
plt.ylabel("Number of Hotels", size=15)
plt.title("Number of Hotels per city", size=19)
# plt.show()
plt.savefig("Visualization/hotels_city_bar.png",
            bbox_inches ="tight", 
            pad_inches = 0.5, 
            transparent = True, 
            facecolor ="w", 
            edgecolor ='b', 
            orientation ='landscape')
```

```python
prices = [df1['Price'], df2['Price'], df3['Price'], df4['Price'], df5['Price']]
f = plt.figure(figsize=(10,10))
a = f.add_subplot(111)
a.set_xticklabels(['Bangalore', 'Delhi', 'Mumbai', 'Chennai', 'Kolkata'])
a.set_title('Hotel prices (in Rupees) accross cities', size=17)
plt.boxplot(prices)
# plt.show()
plt.savefig("Visualization/box_whisker.png",
            bbox_inches ="tight", 
            pad_inches = 0.5, 
            transparent = True, 
            facecolor ="w", 
            edgecolor ='b')
```
