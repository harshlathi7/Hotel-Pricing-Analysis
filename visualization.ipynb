```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from scipy.stats import norm
```

```python
df1 = pd.DataFrame(pd.read_csv("Dataset/Cleaned Data/bangalore_hotels_cleaned.csv"))
df2 = pd.DataFrame(pd.read_csv("Dataset/Cleaned Data/delhi_hotels_cleaned.csv"))
df3 = pd.DataFrame(pd.read_csv("Dataset/Cleaned Data/mumbai_hotels_cleaned.csv"))
df4 = pd.DataFrame(pd.read_csv("Dataset/Cleaned Data/chennai_hotels_cleaned.csv"))
df5 = pd.DataFrame(pd.read_csv("Dataset/Cleaned Data/kolkata_hotels_cleaned.csv"))
```

```python
x = ['Bangalore', 'Delhi', 'Mumbai', 'Chennai', 'Kolkata']
y = [df1.shape[0], df2.shape[0], df3.shape[0], df4.shape[0], df5.shape[0]]

df = pd.DataFrame({"City":x, "Hotels":y})
df = df.sort_values('Hotels', ascending=False)

plt.figure(figsize=(10,6))
plt.bar('City', 'Hotels', data=df)
plt.xlabel("City", size=15)
plt.ylabel("Number of Hotels", size=15)
plt.show()
```
